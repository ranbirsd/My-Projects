But here 12am means the date changes to the next the day (2022-10-17- which is not required here) so we will take the time to the last minute of the 11th hour from 12pm that is 23:59:00.

```python
# Reading User GamePlay Data File, Deposit Data File and Withdrawal Data File.

b1_ugpd = pd.read_csv("User Gameplay Data.csv")
b1_dd = pd.read_csv("Deposit Data.csv")
b1_wd = pd.read_csv("Withdrawal Data.csv")

# Changing the data type for Datetime, DDt and WDt columns (variables) to datetime64[ns]

b1_ugpd['Datetime'] = b1_ugpd['Datetime'].astype('datetime64[ns]')
b1_dd['DDt'] = b1_dd['DDt'].astype('datetime64[ns]')
b1_wd['WDt'] = b1_wd['WDt'].astype('datetime64[ns]')


# Extracting the required datetime slot- 16th October 2022 12pm to 12am (i.e; 23:59:00) for 
# Datetime, DDt and WDt columns in User Gameplay Data, Deposit Data and Withdrawal Data datasets respectively.

b1_ugpdf = b1_ugpd[(b1_ugpd["Datetime"]>="2022-10-16 12:00:00") &(b1_ugpd["Datetime"]<="2022-10-16 23:59:00")]
b1_ddf = b1_dd[(b1_dd["DDt"]>="2022-10-16 12:00:00") &(b1_dd["DDt"]<="2022-10-16 23:59:00")]
b1_wdf = b1_wd[(b1_wd["WDt"]>="2022-10-16 12:00:00") &(b1_wd["WDt"]<="2022-10-16 23:59:00")]
```

```python
# Looking number of rows and columns for all the datasets

b1_ugpdf.shape, b1_ddf.shape, b1_wdf.shape
```

```python
# displaying first 5 rows of all the dataframes

b1_ugpdf.head(), b1_ddf.head(), b1_wdf.head()
```

```python
# Mutliplying 0.01 in the Deposit Amount (DA) column of the Deposit Data Dataset as required to calculate the Loyalty Points.
# Mutliplying 0.005 in the Withdrawal Amount (WA) column of the Withdrawal Data Dataset as required to calculate Loyalty Points.
# Mutliplying 0.2 in the GamesPlayed column of the User Gameplay Data Dataset as required to calculate the Loyalty Points.
# making new columns in all 3 dataframes.

b1_ddf["ddflp"] = 0.01* b1_ddf["DA"] 
b1_wdf["wdflp"] = 0.005* b1_wdf["WA"] 
b1_ugpdf["ugpdflp"] = 0.2* b1_ugpdf["GamesPlayed"]
```

```python
# Grouping User Id of Deposit Data dataset by count for the required slot to find the number of deposits done by user.
# And then storing it in a new variable for count.
# Dropping the columns which are not required as we only need number of deposits done by a user.


b1_ddf_count = b1_ddf.groupby(["UserId"]).count()
b1_ddf_count.drop(['DDt', 'ddflp'], axis=1, inplace=True)

# Grouping User Id of Withdrawal Data dataset by count for the required slot to find the number of withdrawals done by user.
# And then storing it in a new variable for count.
# Dropping the columns which are not required as we only need number of withdrawals done by a user.


b1_wdf_count = b1_wdf.groupby(["UserId"]).count()
b1_wdf_count.drop(['WDt', 'wdflp'], axis=1, inplace=True)
```

```python
# Merging dataframes by using Left Join on count dataframes of deposit data dataset 
# and withdrawal data dataset for the required datetime slot.
# Displaying first 5 rows for the merged dataframe.

b1_mrg = pd.merge(b1_ddf_count, b1_wdf_count, on = "UserId", how = "left")
b1_mrg.head()
```

```python
# making null values = 0.
# displaying first 5 rows

b1_mrg.loc[(b1_mrg.WA.isna() == True) , 'WA'] = 0 
b1_mrg.head()
```

```python
# looking for data type of Withdrawal Amount (WA) column

b1_mrg["WA"].dtypes
```

```python
# changing the data type of Withdrawal Amount from float64 to int64

b1_mrg['WA'] = b1_mrg['WA'].astype('int64')
```

```python
# Making a new column for calculating Loyalty Points of 
# 0.001 * (max of ((number of deposit amount - number of withdrawal amount) or 0)).

b1_mrg["d-w_lp"] = 0.001* (b1_mrg["DA"]- b1_mrg["WA"])
```

```python
# making the negative values of the d-w_lp column = 0, as we only need the maximum value of the difference
# of the number deposit amount and the number of withdrawal amount, or 0. We don't need negative values.

# making the negative values = 0, as we only need the maximum value from the substraction of DA - WA.
# that is DA - WA is not less than 0 but greater than or equal to 0.

b1_mrg.loc[(b1_mrg["d-w_lp"] < 0) , 'd-w_lp'] = 0 
```

```python
# Grouping User Id of Withdrawal Data dataset by sum for the required slot to find the sum of withdrawal done by user.
# And then storing it in a new variable for sum.

# Grouping User Id of Deposit Data dataset by sum for the required slot to find the sum of deposits done by user.
# And then storing it in a new variable for sum.

# Grouping User Id of User Gameplay Data dataset by sum of GamesPlayed and user gameplay data loyalty point 
# for the required slot to find the sum of GamesPlayed by user.
# And then storing it in a new variable for sum.

b1_wdf_sum = b1_wdf.groupby(["UserId"]).sum()
b1_ddf_sum = b1_ddf.groupby(["UserId"]).sum()
b1_ugpdf_sum = b1_ugpdf.groupby(["UserId"]).sum(["GamesPlayed", "ugpdlp"])
```
