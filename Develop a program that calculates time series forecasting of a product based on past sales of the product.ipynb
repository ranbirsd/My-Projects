# c.	Last Year to This Year

```python
# function to perform last year to this year method
def lastYearToThisYear(df):
    last_year = df['Sales of Bath Towel'].iloc[-12:]
    this_year = df['Sales of Bath Towel'].iloc[-6:]
    forecast = this_year.mean() + (this_year.mean() - last_year.mean())
    forecast = [forecast for i in range(6)]
    return forecast
```

# j.	Exponential Smoothing

```python
# function to perform exponential smoothing method
def exponentialSmoothing(df):
    forecasts = []
    alpha = 0.5
    for i in range(len(df['Sales of Bath Towel']) + 6):
        if i == 0:
            forecasts.append(df['Sales of Bath Towel'].iloc[0])
        elif i < len(df['Sales of Bath Towel']):
            forecasts.append(df['Sales of Bath Towel'].iloc[i])
        else:
            forecast = alpha * df['Sales of Bath Towel'].iloc[-1] + (1 - alpha) * forecasts[-1]
            forecasts.append(forecast)
    return forecasts
```

# k.	Exponential Smoothing with Trend and Seasonality

```python
# function to perform exponential smoothing with trend and seasonality method
def exponentialSmoothingWithTrendAndSeasonality(df):
    forecasts = []
    alpha = 0.5
    beta = 0.5
    gamma = 0.5
    for i in range(len(df['Sales of Bath Towel']) + 6):
        if i == 0:
            forecasts.append(df['Sales of Bath Towel'].iloc[0])
        elif i == 1:
            forecasts.append(df['Sales of Bath Towel'].iloc[1])
        elif i < len(df['Sales of Bath Towel']):
            forecasts.append(df['Sales of Bath Towel'].iloc[i])
        else:
            level = alpha * df['Sales of Bath Towel'].iloc[-1] + (1 - alpha) * (forecasts[-1] + forecasts[-2]) / 2
            trend = beta * (level - forecasts[-1]) + (1 - beta) * (forecasts[-1] - forecasts[-2]) / 2
            seasonality = gamma * (df['Sales of Bath Towel'].iloc[-1] - level) + (1 - gamma) * (forecasts[-1] / level)
            forecast = level + trend + seasonality
            forecasts.append(forecast)
    return forecasts
```
