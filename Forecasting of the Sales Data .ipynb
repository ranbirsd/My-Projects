# f). Least Square Regression

```python
import statsmodels.api as sm

def least_square_regression_forecast(sales_data, window_size):
    X = np.array(range(len(sales_data)-window_size, len(sales_data)))
    y = sales_data[-window_size:]
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()
    X_forecast = np.array(range(len(sales_data), len(sales_data) + 6))
    X_forecast = sm.add_constant(X_forecast)
    y_forecast = model.predict(X_forecast)
    return y_forecast

# Calculate the forecast sales values for each product using the least_square_regression_forecast function
bath_towel_forecast = least_square_regression_forecast(bath_towel_sales, 6)
bedsheet_forecast = least_square_regression_forecast(bedsheet_sales, 6)
pillow_forecast = least_square_regression_forecast(pillow_sales, 6)
curtains_forecast = least_square_regression_forecast(curtains_sales, 6)
umbrella_forecast = least_square_regression_forecast(umbrella_sales, 6)

# Create a DataFrame to store the forecast sales data
forecast6_df = pd.DataFrame({'SKU': ['Bath Towel', 'Bedsheet', 'Pillow', 'Curtains', 'Umbrella'],
                            'Jan 2023': [bath_towel_forecast[0], bedsheet_forecast[0], pillow_forecast[0], curtains_forecast[0], umbrella_forecast[0]],
'Feb 2023': [bath_towel_forecast[1], bedsheet_forecast[1], pillow_forecast[1], curtains_forecast[1], umbrella_forecast[1]],
'Mar 2023': [bath_towel_forecast[2], bedsheet_forecast[2], pillow_forecast[2], curtains_forecast[2], umbrella_forecast[2]],
'Apr 2023': [bath_towel_forecast[3], bedsheet_forecast[3], pillow_forecast[3], curtains_forecast[3], umbrella_forecast[3]],
'May 2023': [bath_towel_forecast[4], bedsheet_forecast[4], pillow_forecast[4], curtains_forecast[4], umbrella_forecast[4]],
'Jun 2023': [bath_towel_forecast[5], bedsheet_forecast[5], pillow_forecast[5], curtains_forecast[5], umbrella_forecast[5]]})

# Set the index of the DataFrame to be the 'SKU' column
forecast6_df.set_index('SKU', inplace=True)

# Add a title to the DataFrame
forecast6_df.title = "Forecast Sales using Least Square Regression Method"

# Print the resulting DataFrame
print(forecast6_df)

# Save the forecast sales data as a csv file
forecast6_df.to_csv('Forecasting of the Sales Data using Least Square Regression Method.csv', index=True)

#reading and writing file again and adding the title in the csv
title = "Forecast Sales using Least Square Regression Method"
with open("Forecasting of the Sales Data using Least Square Regression Method.csv", "w") as f:
    f.write(title + "\n")
    forecast6_df.to_csv(f, index=True, header="SKU")
```

# g). Second Degree Approximation

```python
def second_degree_approximation_forecast(sales_data, window_size):
    X = np.array(range(len(sales_data)-window_size, len(sales_data)))
    y = sales_data[-window_size:]
    coeffs = np.polyfit(X, y, 2)
    poly = np.poly1d(coeffs)
    X_forecast = np.array(range(len(sales_data), len(sales_data) + 6))
    y_forecast = poly(X_forecast)
    return y_forecast

# Calculate the forecast sales values for each product using the second_degree_approximation_forecast function
bath_towel_forecast = second_degree_approximation_forecast(bath_towel_sales, 6)
bedsheet_forecast = second_degree_approximation_forecast(bedsheet_sales, 6)
pillow_forecast = second_degree_approximation_forecast(pillow_sales, 6)
curtains_forecast = second_degree_approximation_forecast(curtains_sales, 6)
umbrella_forecast = second_degree_approximation_forecast(umbrella_sales, 6)

# Create a DataFrame to store the forecast sales data
forecast7_df = pd.DataFrame({'SKU': ['Bath Towel', 'Bedsheet', 'Pillow', 'Curtains', 'Umbrella'],
                            'Jan 2023': [bath_towel_forecast[0], bedsheet_forecast[0], pillow_forecast[0], curtains_forecast[0], umbrella_forecast[0]],
'Feb 2023': [bath_towel_forecast[1], bedsheet_forecast[1], pillow_forecast[1], curtains_forecast[1], umbrella_forecast[1]],
'Mar 2023': [bath_towel_forecast[2], bedsheet_forecast[2], pillow_forecast[2], curtains_forecast[2], umbrella_forecast[2]],
'Apr 2023': [bath_towel_forecast[3], bedsheet_forecast[3], pillow_forecast[3], curtains_forecast[3], umbrella_forecast[3]],
'May 2023': [bath_towel_forecast[4], bedsheet_forecast[4], pillow_forecast[4], curtains_forecast[4], umbrella_forecast[4]],
'Jun 2023': [bath_towel_forecast[5], bedsheet_forecast[5], pillow_forecast[5], curtains_forecast[5], umbrella_forecast[5]]})

# Set the index of the DataFrame to be the 'SKU' column
forecast7_df.set_index('SKU', inplace=True)

# Add a title to the DataFrame
forecast7_df.title = "Forecast Sales using Second Degree Approximation Method"

# Print the resulting DataFrame
print(forecast7_df)

# Save the forecast sales data as a csv file
forecast7_df.to_csv('Forecasting of the Sales Data using Second Degree Approximation Method.csv', index=True)

#reading and writing file again and adding the title in the csv
title = "Forecast Sales using Second Degree Approximation Method"
with open("Forecasting of the Sales Data using Second Degree Approximation Method.csv", "w") as f:
    f.write(title + "\n")
    forecast7_df.to_csv(f, index=True, header="SKU")
```

# h). Weighted Moving Average

```python
def weighted_moving_average_forecast(sales_data, window_size):
    weights = np.exp(np.linspace(-1., 0., window_size))
    weights /= weights.sum()
    y_forecast = np.convolve(sales_data, weights, mode='valid')[-window_size:]
    return y_forecast

# Calculate the forecast sales values for each product using the weighted_moving_average_forecast function
bath_towel_forecast = weighted_moving_average_forecast(bath_towel_sales, 6)
bedsheet_forecast = weighted_moving_average_forecast(bedsheet_sales, 6)
pillow_forecast = weighted_moving_average_forecast(pillow_sales, 6)
curtains_forecast = weighted_moving_average_forecast(curtains_sales, 6)
umbrella_forecast = weighted_moving_average_forecast(umbrella_sales, 6)

# Create a DataFrame to store the forecast sales data
forecast8_df = pd.DataFrame({'SKU': ['Bath Towel', 'Bedsheet', 'Pillow', 'Curtains', 'Umbrella'],
                            'Jan 2023': [bath_towel_forecast[0], bedsheet_forecast[0], pillow_forecast[0], curtains_forecast[0], umbrella_forecast[0]],
'Feb 2023': [bath_towel_forecast[1], bedsheet_forecast[1], pillow_forecast[1], curtains_forecast[1], umbrella_forecast[1]],
'Mar 2023': [bath_towel_forecast[2], bedsheet_forecast[2], pillow_forecast[2], curtains_forecast[2], umbrella_forecast[2]],
'Apr 2023': [bath_towel_forecast[3], bedsheet_forecast[3], pillow_forecast[3], curtains_forecast[3], umbrella_forecast[3]],
'May 2023': [bath_towel_forecast[4], bedsheet_forecast[4], pillow_forecast[4], curtains_forecast[4], umbrella_forecast[4]],
'Jun 2023': [bath_towel_forecast[5], bedsheet_forecast[5], pillow_forecast[5], curtains_forecast[5], umbrella_forecast[5]]})

# Set the index of the DataFrame to be the 'SKU' column
forecast8_df.set_index('SKU', inplace=True)

# Add a title to the DataFrame
forecast8_df.title = "Forecast Sales using Weighted Moving Average Method"

# Print the resulting DataFrame
print(forecast8_df)

# Save the forecast sales data as a csv file
forecast8_df.to_csv('Forecasting of the Sales Data using Weighted Moving Average Method.csv', index=True)

#reading and writing file again and adding the title in the csv
title = "Forecast Sales using Weighted Moving Average Method"
with open("Forecasting of the Sales Data using Weighted Moving Average Method.csv", "w") as f:
    f.write(title + "\n")
    forecast8_df.to_csv(f, index=True, header="SKU")
```

# i). Linear Smoothing

```python
def linear_smoothing_forecast(sales_data, alpha):
    y_forecast = [sales_data[0]]
    for i in range(1, len(sales_data)):
        y_forecast.append(alpha * sales_data[i] + (1 - alpha) * y_forecast[i-1])
    return y_forecast[-6:]

# Calculate the forecast sales values for each product using the linear_smoothing_forecast function
bath_towel_forecast = linear_smoothing_forecast(bath_towel_sales, 0.5)
bedsheet_forecast = linear_smoothing_forecast(bedsheet_sales, 0.5)
pillow_forecast = linear_smoothing_forecast(pillow_sales, 0.5)
curtains_forecast = linear_smoothing_forecast(curtains_sales, 0.5)
umbrella_forecast = linear_smoothing_forecast(umbrella_sales, 0.5)

# Create a DataFrame to store the forecast sales data
forecast9_df = pd.DataFrame({'SKU': ['Bath Towel', 'Bedsheet', 'Pillow', 'Curtains', 'Umbrella'],
'Jan 2023': [bath_towel_forecast[0], bedsheet_forecast[0], pillow_forecast[0], curtains_forecast[0], umbrella_forecast[0]],
'Feb 2023': [bath_towel_forecast[1], bedsheet_forecast[1], pillow_forecast[1], curtains_forecast[1], umbrella_forecast[1]],
'Mar 2023': [bath_towel_forecast[2], bedsheet_forecast[2], pillow_forecast[2], curtains_forecast[2], umbrella_forecast[2]],
'Apr 2023': [bath_towel_forecast[3], bedsheet_forecast[3], pillow_forecast[3], curtains_forecast[3], umbrella_forecast[3]],
'May 2023': [bath_towel_forecast[4], bedsheet_forecast[4], pillow_forecast[4], curtains_forecast[4], umbrella_forecast[4]],
'Jun 2023': [bath_towel_forecast[5], bedsheet_forecast[5], pillow_forecast[5], curtains_forecast[5], umbrella_forecast[5]]})

# Set the index of the DataFrame to be the 'SKU' column
forecast9_df.set_index('SKU', inplace=True)

# Add a title to the DataFrame
forecast9_df.title = "Forecast Sales using Linear Smoothing Method"

# Print the resulting DataFrame
print(forecast9_df)

# Save the forecast sales data as a csv file
forecast9_df.to_csv('Forecasting of the Sales Data using Linear Smoothing Method.csv', index=True)

#reading and writing file again and adding the title in the csv
title = "Forecast Sales using Linear Smoothing Method"
with open("Forecasting of the Sales Data using Linear Smoothing Method.csv", "w") as f:
    f.write(title + "\n")
    forecast9_df.to_csv(f, index=True, header="SKU")
```

```python
def exponential_smoothing_forecast(sales_data, alpha):
    y_forecast = [sales_data[0]]
    for i in range(1, len(sales_data)):
        y_forecast.append(alpha * sales_data[i] + (1 - alpha) * y_forecast[i-1])
    return y_forecast[-6:]

# Calculate the forecast sales values for each product using the exponential_smoothing_forecast function
bath_towel_forecast = exponential_smoothing_forecast(bath_towel_sales, 0.5)
bedsheet_forecast = exponential_smoothing_forecast(bedsheet_sales, 0.5)
pillow_forecast = exponential_smoothing_forecast(pillow_sales, 0.5)
curtains_forecast = exponential_smoothing_forecast(curtains_sales, 0.5)
umbrella_forecast = exponential_smoothing_forecast(umbrella_sales, 0.5)

# Create a DataFrame to store the forecast sales data
forecast10_df = pd.DataFrame({'SKU': ['Bath Towel', 'Bedsheet', 'Pillow', 'Curtains', 'Umbrella'],
'Jan 2023': [bath_towel_forecast[0], bedsheet_forecast[0], pillow_forecast[0], curtains_forecast[0], umbrella_forecast[0]],
'Feb 2023': [bath_towel_forecast[1], bedsheet_forecast[1], pillow_forecast[1], curtains_forecast[1], umbrella_forecast[1]],
'Mar 2023': [bath_towel_forecast[2], bedsheet_forecast[2], pillow_forecast[2], curtains_forecast[2], umbrella_forecast[2]],
'Apr 2023': [bath_towel_forecast[3], bedsheet_forecast[3], pillow_forecast[3], curtains_forecast[3], umbrella_forecast[3]],
'May 2023': [bath_towel_forecast[4], bedsheet_forecast[4], pillow_forecast[4], curtains_forecast[4], umbrella_forecast[4]],
'Jun 2023': [bath_towel_forecast[5], bedsheet_forecast[5], pillow_forecast[5], curtains_forecast[5], umbrella_forecast[5]]})

# Set the index of the DataFrame to be the 'SKU' column
forecast10_df.set_index('SKU', inplace=True)

# Add a title to the DataFrame
forecast10_df.title = "Forecast Sales using Exponential Smoothing Method"

# Print the resulting DataFrame
print(forecast10_df)

# Save the forecast sales data as a csv file
forecast10_df.to_csv('Forecasting of the Sales Data using Exponential Smoothing Method.csv', index=True)

#reading and writing file again and adding the title in the csv
title = "Forecast Sales using Exponential Smoothing Method"
with open("Forecasting of the Sales Data using Exponential Smoothing Method.csv", "w") as f:
    f.write(title + "\n")
    forecast10_df.to_csv(f, index=True, header="SKU")
```

# k). Exponential Smoothing with Trend and Seasonality

```python
# To implement exponential smoothing with trend and seasonality, we need to implement Holt-Winters method, 
# which involves updating two additional smoothed values (one for the trend and one for the seasonality) 
# along with the forecast value.

def holt_winters_forecast(sales_data, alpha, beta, gamma, m):
    y_forecast = [sales_data[0]]
    t_forecast = [sales_data[1] - sales_data[0]]
    s_forecast = [sales_data[0]] * m
    for i in range(1, len(sales_data)):
        y_t = alpha * (sales_data[i] - s_forecast[(i-1) % m]) + (1 - alpha) * (y_forecast[i-1] + t_forecast[i-1])
        t_t = beta * (y_t - y_forecast[i-1]) + (1 - beta) * t_forecast[i-1]
        s_t = gamma * (sales_data[i] - y_t) + (1 - gamma) * s_forecast[(i-1) % m]
        y_forecast.append(y_t + t_t + s_t)
        t_forecast.append(t_t)
        s_forecast.append(s_t)
    return y_forecast[-6:]

# Calculate the forecast sales values for each product using the holt_winters_forecast function
bath_towel_forecast = holt_winters_forecast(bath_towel_sales, 0.5, 0.5, 0.5, 12)
bedsheet_forecast = holt_winters_forecast(bedsheet_sales, 0.5, 0.5, 0.5, 12)
pillow_forecast = holt_winters_forecast(pillow_sales, 0.5, 0.5, 0.5, 12)
curtains_forecast = holt_winters_forecast(curtains_sales, 0.5, 0.5, 0.5, 12)
umbrella_forecast = holt_winters_forecast(umbrella_sales, 0.5, 0.5, 0.5, 12)

# Create a DataFrame to store the forecast sales data
forecast11_df = pd.DataFrame({'SKU': ['Bath Towel', 'Bedsheet', 'Pillow', 'Curtains', 'Umbrella'],
'Jan 2023': [bath_towel_forecast[0], bedsheet_forecast[0], pillow_forecast[0], curtains_forecast[0], umbrella_forecast[0]],
'Feb 2023': [bath_towel_forecast[1], bedsheet_forecast[1], pillow_forecast[1], curtains_forecast[1], umbrella_forecast[1]],
'Mar 2023': [bath_towel_forecast[2], bedsheet_forecast[2], pillow_forecast[2], curtains_forecast[2], umbrella_forecast[2]],
'Apr 2023': [bath_towel_forecast[3], bedsheet_forecast[3], pillow_forecast[3], curtains_forecast[3], umbrella_forecast[3]],
'May 2023': [bath_towel_forecast[4], bedsheet_forecast[4], pillow_forecast[4], curtains_forecast[4], umbrella_forecast[4]],
'Jun 2023': [bath_towel_forecast[5], bedsheet_forecast[5], pillow_forecast[5], curtains_forecast[5], umbrella_forecast[5]]})

# Set the index of the DataFrame to be the 'SKU' column
forecast11_df.set_index('SKU', inplace=True)

# Add a title to the DataFrame
forecast11_df.title = "Forecast Sales using Exponential Smoothing with Trend and Seasonality Method"

# Print the resulting DataFrame
print(forecast11_df)

# Save the forecast sales data as a csv file
forecast11_df.to_csv('Forecasting of the Sales Data using Exponential Smoothing with Trend and Seasonality Method.csv', index=True)

# Reading and writing the file again and adding the title in the csv
title = "Forecast Sales using Exponential Smoothing with Trend and Seasonality Method"
with open("Forecasting of the Sales Data using Exponential Smoothing with Trend and Seasonality Method.csv", "w") as f:
    f.write(title + "\n")
    forecast11_df.to_csv(f, index=True, header="SKU")
```
